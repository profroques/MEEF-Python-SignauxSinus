<h2 style = "color : brown">Exemple 1.a</h2>
Le script suivant exécute l'élévation au carré de chaque élément d'une séquence. Sa rédaction repose essentiellement sur l'usage d'une liste.

```python
"""
S. Roques - Académie de Lyon - 2023
"""
# Utilisation d'une liste pour stocker les données
ma_liste = [1, 2, 3, 4, 5]

# Élévation au carré de chaque élément en utilisant une boucle
resultat_liste = []
for element in ma_liste:
    resultat_liste.append(element ** 2)
print("Séquence initiale : ", ma_liste)
print("Résultat avec une liste :", resultat_liste)
```

<h2 style = "color : brown">Exemple 1.b</h2>
Le script suivant exécute l'élévation au carré de chaque élément d'une séquence. Sa rédaction repose essentiellement sur l'usage d'un Numpy Array. 

```python
"""
S. Roques - Académie de Lyon - 2023
"""

import numpy as np

# Utilisation d'un tableau numpy pour stocker les données
mon_tableau = np.array([1, 2, 3, 4, 5])

# Élévation au carré de chaque élément en utilisant des opérations vectorisées
resultat_tableau = mon_tableau ** 2

print("La séquence initiale : ", mon_tableau)
print("Résultat avec un tableau numpy :", resultat_tableau)
```

<h2 style = "color : brown">Exemple 2.a</h2>
Un exemple de script utilisant des listes et permettant d'afficher la représentation graphique y = cos(2.x) avant et après la modification d'une seule valeur de x

```python
"""
S. Roques - Académie de Lyon - 2023
"""

import math
import matplotlib.pyplot as plt

# Création d'une liste pour les valeurs de x
x_list = []
for i in range(0, 21):
    x = i / 10
    x_list.append(x)  # Liste des valeurs de x de 0 à 2 avec un pas de 0.1

# Calcul des valeurs de y = cos(2x) à l'aide d'une boucle for
y_list = []
for x in x_list:
    y = math.cos(2 * x)
    y_list.append(y)

# Création d'une figure avec un sous-graphique
fig, ax = plt.subplots(figsize=(8, 6))

# Représentation graphique avant modification (bleu en trait continu)
ax.plot(x_list, y_list, 'b-', label='Courbe bleue (avant modification)')
ax.set_title("Représentation graphique de y = cos(2x) - AVEC LISTES")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend()

# Modification d'une valeur
y_list[5] = 0.5  # Modification de la valeur à l'indice 5

# Affichage des valeurs après modification
print("Après modification :")
for i in range(len(x_list)):
    x = x_list[i]
    y = y_list[i]
    print(f"x = {x:.1f}, y = {y:.4f}")

# Représentation graphique après modification (rouge en tirets)
ax.plot(x_list, y_list, 'r--', label='Courbe rouge (après modification)')
ax.legend()

# Affichage du sous-graphique avec les deux courbes
plt.show()
```

<h2 style = "color : brown">Exemple 2.b</h2>
Un exemple de script utilisant des NumpyArray et permettant d'afficher la représentation graphique y = cos(2.x) avant et après la modification d'une seule valeur de x

```python
"""
S. Roques - Académie de Lyon - 2023
"""

import numpy as np
import matplotlib.pyplot as plt

# Création d'un numpy array pour les valeurs de x
x_array = np.arange(0, 2.1, 0.1)  # Array des valeurs de x de 0 à 2 avec un pas de 0.1

# Calcul des valeurs de y = cos(2x)
y_array = np.cos(2 * x_array)

# Création d'une figure avec un sous-graphique
fig, ax = plt.subplots(figsize=(8, 6))

# Représentation graphique avant modification (bleu en trait continu)
ax.plot(x_array, y_array, 'b-', label='Courbe bleue (avant modification)')
ax.set_title("Représentation graphique de y = cos(2x) - AVEC ARRAY")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend()

# Modification d'une valeur
y_array[5] = 0.5  # Modification de la valeur à l'indice 5

# Affichage des valeurs après modification
print("Après modification :")
for x, y in zip(x_array, y_array):
    print(f"x = {x:.1f}, y = {y:.4f}")

# Représentation graphique après modification (rouge en tirets)
ax.plot(x_array, y_array, 'r--', label='Courbe rouge (après modification)')
ax.legend()

# Affichage du sous-graphique avec les deux courbes
plt.show()
```
